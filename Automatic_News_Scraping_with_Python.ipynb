```python
import feedparser
from newspaper import Article
import json

def parse_rss_feed(feed_url):
    feed = feedparser.parse(feed_url)
    articles = []

    for entry in feed.entries:
        article_info = {
            'rss_title': entry.get('title', 'No title'),
            'link': entry.get('link', 'No link'),
            'rss_published': entry.get('published', 'No publish date')
        }
        articles.append(article_info)
    return articles

def extract_article_info(url):
    article = Article(url)
    article.download()
    article.parse()
    return {
        'title': article.title,
        'author': article.authors,
        'publish_date': article.publish_date.isoformat() if article.publish_date else None,
        'text': article.text
    }

def get_articles_from_rss(feed_urls):
    all_articles = []
    for feed_url in feed_urls:
        articles = parse_rss_feed(feed_url)
        for article in articles:
            try:
                article_info = extract_article_info(article['link'])
                article_info['rss_title'] = article['rss_title']
                article_info['rss_published'] = article['rss_published']
                all_articles.append(article_info)
            except Exception as e:
                print(f"Failed to process article at {article['link']}: {e}")
    return all_articles

if __name__ == "__main__":
    rss_feed_urls = [
        "https://www.animenewsnetwork.com/news/rss.xml",
        "https://www.crunchyroll.com/rss",
        "https://www.espn.com/espn/rss/news",
        "http://feeds.bbci.co.uk/sport/rss.xml?edition=uk",
        "http://www.skysports.com/rss/12040"
    ]

    articles = get_articles_from_rss(rss_feed_urls)

    # Save the scraped data to a JSON file
    with open('scraped_articles.json', 'w') as f:
        json.dump(articles, f, indent=4)

    # Print the scraped articles
    for article in articles:
        print(f"RSS Title: {article['rss_title']}")
        print(f"RSS Published: {article['rss_published']}")
        print(f"Title: {article['title']}")
        print(f"Author: {article['author']}")
        print(f"Publish Date: {article['publish_date']}")
        print(f"Text: {article['text'][:200]}...")
        print("-" * 80)
```
